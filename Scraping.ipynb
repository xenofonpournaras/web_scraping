```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
import json

url = "https://www.plaisio.gr/anavathmisi-diktia/anavathmisi-pc/koutia-desktop"
page = requests.get(url) # http get request
page # status 200 connection made successfully
```

```python
page.text
```

```python
soup = BeautifulSoup(page.text, "html.parser") #call the beautifulsoup for html.parser
soup # we can see the html code of the page
```

```python
df = pd.DataFrame(columns = ['URL','Product Name', 'SKU','Brand','Price','Availability', 'Initial Price','EAN'])
df
```

#### We can see from the source code that the products for the chosen category are ordered in li tags, class = "column item" and with a unique id

```python
column_items = soup.find_all("li", attrs={"class": "column item"})
column_items
```

```python
pname = []
pcode = []
pbrand = []
pfprice = []
pprice = []

for column_item in column_items:
    pname.append(column_item.find("span", attrs={"itemprop": "name"}).text) # from html found attribute itemprop with value = name
    pcode.append(column_item.find("span", attrs={"itemprop": "sku"}).text) # from html found attribute itemprop with value = sku
    pbrand.append(column_item.find("span", attrs={"itemprop": "name"}).text.split(' ')[0]) # take first string from the product name
    pfprice.append(column_item.find("div", attrs={"class": "from-price"}).text) # from html found class from-price
    pprice.append(column_item.find("div", attrs={"class": "from-price"}).next_sibling.text)
    
df['Product Name'] = pname
df['SKU'] = pcode
df['Brand'] = pbrand
df['Price'] = pprice
df['Initial Price'] = pfprice
```

```python
df
```

### For the url we parse the href tags, but we want the unique url tags that apply to the products. So we have to create a new soup for that purpose

```python
items = soup.find_all("div", attrs={"class": "image"})
items
```

```python
purls = [] # list for the urls

for item in items[:-2]: #exlude the last value which is the "#"
    for x in item.find_all('a', href=True):
        purls.append('https://www.plaisio.gr'+x['href'])
purls
```

```python
df['URL'] = purls
df
```

## Availability

### We can't parse the availability from the HTML tag because the data for the availability comes from a JS script. So we have to parse the script, find the data in JSON format, exploit the json library (with Beautifoulsoup) in python in order to select the availability

```python
url = "https://www.plaisio.gr/anavathmisi-diktia/anavathmisi-pc/koutia-desktop"
r = requests.get(url)
print(r.status_code)
new_soup = BeautifulSoup(r.content, 'html.parser')
new_soup.title # test
```

### Logic: 1. Inspect source code of page 2. Find the script that has the availability (it was the 18th script tag) 3. Strip the data from the tag in order to have a JSON format file (JSON format parser was used)

```python
script = new_soup.find_all('script')[18].text.strip()[:]
script
```

```python
data = json.loads(script)
data
```

```python
availability = [] #list for availability values

for i in range(len(data['itemListElement']))[:-1]:
    # split string and take the last part which shows the status of the availability
    availability.append(data['itemListElement'][i]['item']['offers']['availability'].split('/')[3])
    
availability
```

```python
df['Availability'] = availability
df
```

### EAN

```python
df_ean = df
df_ean
```

```python
urls_ean = [] #list with urls for products

for i in range(len(df_ean)): # iterate through dataframe
    urls_ean.append(df_ean.iloc[i, 0])

urls_ean
```

```python
pean = [] #product_ean list

for i in urls_ean:
    url_test = i #pass the urls iteratively
    k = requests.get(url_test) 
    soup_test = BeautifulSoup(k.content, 'html.parser')
    script_test = soup_test.find_all('script')[19].text.strip()[70:-1] # make the JSON format from the script tag
    soup_test2=BeautifulSoup(script_test[:script_test.find(');')],"html.parser") # exclude javascript syntax in order to have pure JSON format
    soup_test2 = str(soup_test2) #make it string to pass it to json function
    data_test = json.loads(soup_test2)
    pean.append(data_test['data']['productean'])
    
    
pean
```

# FINAL TABLE

```python
df['EAN'] = pean 
df
```
